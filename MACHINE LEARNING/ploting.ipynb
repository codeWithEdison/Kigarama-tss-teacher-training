```python
# for creating plots
import matplotlib.pyplot as plt
# for handling data
import pandas as pd
# read our csv into dataframe
music = pd.read_csv('music.csv')
#music
```

```python
#1. scatter
# create a new figure with spefcied size whidt=8, height=6 inches
plt.figure(figsize=(5,4))
# x-axis = age, y_axix = genre
x= music['age']
y= music['genre']
plt.scatter(x,y)
plt.title('scatter plot : age vs music genre')
plt.xlabel('age of person')
plt.ylabel('music genre')
plt.show()
```

```python
plt.figure(figsize=(4,3))

genre_counts = music['genre'].value_counts()
# genre_counts
x=genre_counts.index
y=genre_counts.values
plt.bar(x,y)
# Add labels and title
plt.xlabel('Music Genre')
plt.ylabel('Number of People')
plt.title('Bar Chart: Number of People by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# BOX PLOT
import seaborn as sns

plt.figure(figsize=(4,3))
sns.boxplot(x="genre", y="age", data=music)
# Add labels and title
plt.xlabel('Music Genre')
plt.ylabel('Age')
plt.title('Box Plot: Age Distribution by Genre')
# Rotate x-axis labels
plt.xticks(rotation=45)
plt.show()
```

```python
#4. HEAT MAP
# convert genre to number plt
plt.figure(figsize=(4,3))
music['genre_number'] = pd.factorize(music['genre'])[0]
# create a correlatoin matrix
correlation = music[['age', 'gender', 'genre_number']].corr()
sns.heatmap(correlation, annot=True)
plt.show()
```
